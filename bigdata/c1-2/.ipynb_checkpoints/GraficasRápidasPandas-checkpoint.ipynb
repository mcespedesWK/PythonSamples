## value_counts()

```python
# El archivo se encuentra en el directorio principal del proyecto
# Retrocedo un directorio (../)
df = pd.read_csv("../iris.csv")
```

```python
# Cuando tenemos una columna con muchos valores, nos permite saber cuales se repite
# Por ejemplo una columna Hombre - Mujeres
# Es decir nos cuenta lo que existe dentro de la coumna de cada cosa.
# Utilizo el ejemplo del Iris dataSet
# Quiero saber cuatos hay de cada especie
df.species.value_counts()
```

```python
# Tambien puedo utilizar el .plot() para una grafica rapida de los resultados
df.species.value_counts().plot(kind='bar', grid = True, color=['green','orange','blue'])
plt.show()
```
