#### Hacer las gr치ficas para Iris Dataset con 랋lotnine

```python
import plotnine as p9

import pandas as pd

df = pd.read_csv('iris.csv')
```

Define las opciones del gr치fico usando mapping y est칠ticas aes, para seleccionar las variables que quieres mostrar en el gr치fico

Empezamos definiendo el conjunto de datos que usaremos, colocaremos los ejes y elegiremos una geometr칤a. 

Por lo tanto, los elementos m칤nimos de cualquier gr치fico son data, 'aes' y 'geom-*':

```python

# tipo de geometr칤a utilizar con la opci칩n "p9.geom_point()"

(p9.ggplot(data=df,
           mapping=p9.aes(x='petal_length', y='petal_width'))
     + p9.geom_point(color='blue')
)

#El s칤mbolo + en el paquete plotnine es particularmente 칰til porque te permite modificar los objetos plotnine existentes. 
# Esto significa que puedes configurar f치cilmente el gr치fico con plantillas y explorar convenientemente diferentes tipos de gr치ficos
```

```python
# Para que cada especie se represente en el gr치ico con un color
# indico que 'color' se defina con la columna 'species'
# Se le pueden agregar muchos m치s elementos con p9, como solo adjuntarlos

(p9.ggplot(data=df,
           mapping=p9.aes(x='petal_length', y='petal_width',
                          color='species'))
    + p9.geom_jitter(alpha=0.2)
)
```

```python
# Puedo realizar un gr치fico en serie 

petal = df.groupby(['petal_length', 'species'])['species'].count()

(p9.ggplot(data=df,
           mapping=p9.aes(x='petal_length',
                          y='petal_width',
                          color='species'))
    + p9.geom_line()
)
```

```python

# Podemos reusar este mismo c칩digo y agregar una faceta con facet_wrap en base a una categor칤a 
# para dividir el gr치fico para cada uno de los grupos

(p9.ggplot(data=df,
           mapping=p9.aes(x='petal_length',
                          y='petal_width',
                          color='species'))
    + p9.geom_point()
    + p9.facet_wrap("species")
)
```
