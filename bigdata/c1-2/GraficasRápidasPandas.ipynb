#### Hacer las gráficas para Iris Dataset con “plotnine”

```python
import plotnine as p9

import pandas as pd

df = pd.read_csv('iris.csv')
```

Define las opciones del gráfico usando mapping y estéticas aes, para seleccionar las variables que quieres mostrar en el gráfico

Empezamos definiendo el conjunto de datos que usaremos, colocaremos los ejes y elegiremos una geometría. 

Por lo tanto, los elementos mínimos de cualquier gráfico son data, 'aes' y 'geom-*':

```python

# tipo de geometría utilizar con la opción "p9.geom_point()"

(p9.ggplot(data=df,
           mapping=p9.aes(x='petal_length', y='petal_width'))
     + p9.geom_point(color='blue')
)

#El símbolo + en el paquete plotnine es particularmente útil porque te permite modificar los objetos plotnine existentes. 
# Esto significa que puedes configurar fácilmente el gráfico con plantillas y explorar convenientemente diferentes tipos de gráficos
```

```python
# Para que cada especie se represente en el gráico con un color
# indico que 'color' se defina con la columna 'species'
# Se le pueden agregar muchos más elementos con p9, como solo adjuntarlos

(p9.ggplot(data=df,
           mapping=p9.aes(x='petal_length', y='petal_width',
                          color='species'))
    + p9.geom_jitter(alpha=0.2)
)
```

```python
# Puedo realizar un gráfico en serie 

petal = df.groupby(['petal_length', 'species'])['species'].count()

(p9.ggplot(data=df,
           mapping=p9.aes(x='petal_length',
                          y='petal_width',
                          color='species'))
    + p9.geom_line()
)
```

```python

# Podemos reusar este mismo código y agregar una faceta con facet_wrap en base a una categoría 
# para dividir el gráfico para cada uno de los grupos

(p9.ggplot(data=df,
           mapping=p9.aes(x='petal_length',
                          y='petal_width',
                          color='species'))
    + p9.geom_point()
    + p9.facet_wrap("species")
)
```
